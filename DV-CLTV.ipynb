### 1- RFM

```python
rfm = pd.DataFrame()
```

```python
rfm["frequency"] = df["NumWebPurchases"] + df["NumCatalogPurchases"] + df["NumStorePurchases"] + df["NumWebVisitsMonth"]
```

```python
rfm
```

```python
rfm["monetary"] = df["MntFishProducts"] + df["MntFruits"] + df["MntGoldProds"] + df["MntMeatProducts"] + df["MntSweetProducts"] + df["MntWines"]
```

```python
rfm["recency"] = df["Recency"]
```

```python
rfm
```

```python
rfm["rfm_freq_score"] = pd.qcut(rfm["frequency"], 5, labels=[1,2,3,4,5])
```

```python
rfm["rfm_recency_score"] = pd.qcut(rfm["recency"], 5, labels=[5,4,3,2,1])
```

```python
rfm["rfm_monetary_score"] = pd.qcut(rfm["monetary"], 5, labels=[1,2,3,4,5])
```

```python
rfm
```

```python
rfm["RF Score"] = rfm["rfm_freq_score"].astype(str) + rfm["rfm_recency_score"].astype(str)
```

```python
rfm
```

```python
# rfm["Monetary"] = df[df.columns[df.columns.str.contains("Mnt")]].sum(axis=1)
```

```python
seg_map = {
    r"[1-2][1-2]": "hibernating",  # birinci elemanında 1 yada 2, ikinci elemanında 1 yada 2 görürsen o isimlendirmeyi yap
    r"[1-2][3-4]": "at_Risk",
    r"[1-2]5": "cant_loose",
    r"3[1-2]": "need_attention",
    r"33": "about_to_sleep",
    r"[3-4][4-5]": "loyal_customers",
    r"41": "promising",  # birinci elemanında 4, ikinci elemanında 1 görürsen o isimlendirmeyi yap
    r"51": "new_customers",
    r"[4-5][2-3]": "potential_loyalists",
    r"5[4-5]": "champions",
}

# RFM_SCORE da değişkenleri seg_map deki değerler ile değiştir ve onlara da dağılım uygula yukarıdaki gibi
rfm["segment"] = rfm["RF Score"].replace(seg_map, regex=True)
```

### CLTV

```python
df["monetary_avg"] = rfm["monetary"] / rfm["frequency"]
```

```python
df["Dt_Customer"].max()
```

```python
cltv = pd.DataFrame()
```

```python
cltv_df = df.groupby("ID").agg({"Dt_Customer":lambda x: (current_Day - x).dt.days,
                               "Recency": lambda x: x,
                               "monetary_avg": lambda x: x,
                              "frequency": lambda x: x})
```

```python
cltv_df.columns = ["T", "Recency", "Monetary", "Frequency"]
```

```python
cltv_df = cltv_df[(cltv_df["Frequency"] > 0)]
```

```python
cltv_df
```

```python
cltv_df["Recency"] = cltv_df["Recency"] / 7
cltv_df["T"] = cltv_df["T"] / 7
```

```python
cltv_df["Recency"].max()
cltv_df["T"].min()
```

```python
cltv_df
```

```python
# cltv_df = cltv_df[cltv_df["T"] > cltv_df["Recency"]]
```

```python
# !pip install lifetimes
```

```python
from lifetimes import BetaGeoFitter
from lifetimes import GammaGammaFitter
from lifetimes.plotting import plot_period_transactions 
```

```python
bgf = BetaGeoFitter(penalizer_coef=0.001)

bgf.fit(cltv_df['Frequency'],
        cltv_df['Recency'],
        cltv_df['T'])
```

### 3 aylık süreçte bizden en çok sipariş vermesini beklediğimiz müşteriler

```python
bgf.predict(12,
        cltv_df['Frequency'],
        cltv_df['Recency'],
        cltv_df['T']).sort_values(ascending=False).head(8)
```

### 4 aylık süreçte bizden en çok sipariş vermesini beklediğimiz müşteriler

```python
bgf.predict(16,
        cltv_df['Frequency'],
        cltv_df['Recency'],
        cltv_df['T']).sort_values(ascending=False).head(8)
```

```python
ggf = GammaGammaFitter(penalizer_coef=0.01)
```

```python
ggf.fit(cltv_df['Frequency'], cltv_df['Monetary'])
```

### Müşteri başına bize bırakacağı parasal değerler

```python
ggf.conditional_expected_average_profit(cltv_df['Frequency'],
                                        cltv_df['Monetary']).sort_values(ascending=False).head(10)
```

```python
cltv = ggf.customer_lifetime_value(bgf,
                                   cltv_df['Frequency'],
                                   cltv_df['Recency'],
                                   cltv_df['T'],
                                   cltv_df['Monetary'],
                                   time=3,  # 3 aylık
                                   freq="W",  # T'nin frekans bilgisi.
                                   discount_rate=0.01)
```

```python
cltv.sort_values(ascending=False)
```

```python
cltv.head()
```

```python
cltv.sort_values( ascending=False).head(10)
```

```python
cltv_final = cltv_df.merge(cltv, on="ID", how="left")
cltv_final.sort_values(by="clv", ascending=False).head(20)
```

```python
cltv_final["segment"] = pd.qcut(cltv_final["clv"], 4, labels=["D", "C", "B", "A"])
```
